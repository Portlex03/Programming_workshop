#### 5.1 гистограмму распределения любого числового признака (используется один числовой столбец);

```python
import matplotlib.pyplot as plt
df.hist('book price',figsize=(7,3))
plt.savefig('hist')
```

#### 5.2 диаграмму «ящик с усами» для любого числового признака (используется один числовой столбец);

```python
df.boxplot('rating',figsize=(7,3))
plt.savefig('boxplot')
```

#### 5.3 круговую диаграмму (выделите номинативный признак, например пол, и используйте количество соответствующих строк);

```python
from collections import Counter
import seaborn as sns

info = dict(Counter(df['rating'].dropna()))
info
```

```python
plt.pie(
    x=list(info.values()), 
    labels=list(info.keys())
)
plt.title('Круговая диаграмма колонки rating')
plt.savefig('plt_pie')
```

#### 5.4 тепловую карту со значениями взаимной корреляции между всеми парами признаков набора данных (значения корреляции должны быть выведены, размер рисунка подобрать так, чтобы было читаемо);

```python
sns.set(rc={'figure.figsize':(7,3)})
dataplot = sns.heatmap(df.corr(numeric_only=True),cmap="YlGnBu", annot=True)
plt.savefig('hot_map')
```

#### 5.5. диаграмму countplot с группировкой по двум номинативным признакам

```python
sns.set(rc={'figure.figsize':(11,8)})
sns.countplot(df.query('year > 2005'),x='rating',hue='year')
plt.savefig('diagram')
```
