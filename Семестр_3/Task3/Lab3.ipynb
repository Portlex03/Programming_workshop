## Задание 1

Взять свою матрицу из 1.10

```python
import numpy as np
M = np.array([[2,  -5,    1,     0],
              [1,  -1,  -13,     0],
              [3,  -2,   -2,    -4],
              [4,   0,  2.7,  -1.3]])
```

## Задание 2

Получить LU-разложение (LUP-разложение) (факторизацию матрицы)

```python
def print_matrix(matrix_name: str, matrix: np.ndarray) -> None:
    print(matrix_name + ':')
    print('\n'.join('|'.join("{:8}".format(str(round(cell,5))) for cell in row) for row in matrix))
    print('\n')
```

```python
from scipy.linalg import lu

P,L,U = lu(M)
print_matrix('P',P)
print_matrix('L',L)
print_matrix('U',U)
```

## Задание 3

Найти определитель (произведение элементов на главной диагонали L и
U, определителя обратной матрицы к P)

```python
print('Определитель матрицы M:',np.linalg.det(M))
print('Определитель матрицы M другим способом:',\
      np.prod(np.diag(L) * np.diag(U)) * np.linalg.det(np.linalg.inv(P)))
```

## Задание 4

Сгенерировать две выборки (вектор целых чисел из 100 элементов) в
одинаковых интервалах с разными распределениями (равномерное и
нормальное)

```python
uniform_aray = np.random.uniform(0,100,size=100)
normal_array = np.linspace(0,100,100)

print('Равномерное распределение:')
print(uniform_aray,'\n')

print('Нормальное распределение:')
print(normal_array)
```

## Задание 5

Вычислить для каждой из выборок:
1. Среднее;
2. Моду;
3. Медиану;
4. Минимум;
5. Максимум;
6. Стандартное отклонение.

```python
from scipy.stats import mode

print('Равномерное распределение:')
print(f'1. Среднее значение: {uniform_aray.mean()}')
print(f'2. Мода: {mode(uniform_aray)[0]}')
print(f'3. Медиана: {np.median(uniform_aray)}')
print(f'3. Минимум: {uniform_aray.min()}')
print(f'5. Максимум: {uniform_aray.max()}')
print(f'6. Стандартное отклонение: {uniform_aray.std()}\n')

print('Нормальное распределение:')
print(f'1. Среднее значение: {normal_array.mean()}')
print(f'2. Мода: {mode(normal_array)[0]}')
print(f'3. Медиана: {np.median(normal_array)}')
print(f'3. Минимум: {normal_array.min()}')
print(f'5. Максимум: {normal_array.max()}')
print(f'6. Стандартное отклонение: {normal_array.std()}')
```

## Задание 6

С помощью метода хи-квадрат (scipy.stats.chisquare) вычислить значение
p-value для нулевой гипотезы: “Распределение выборки не
равномерное”.

```python
from scipy.stats import chisquare

print('p-value:',chisquare(uniform_aray).pvalue)
```
