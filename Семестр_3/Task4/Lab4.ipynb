## –ü—É–Ω–∫—Ç 6

–∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∞—Å—å –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏), –ø–æ—Å—Ç—Ä–æ–π—Ç–µ
—Ç—Ä–µ—Ö–º–µ—Ä–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏. 
- –£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ —Ü–≤–µ—Ç –ª–∏–Ω–∏–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ.
- –í–∞–∂–Ω–æ: –µ—Å–ª–∏ –≤ –≤–∞—à–µ–º –≤–∞—Ä–∏–∞–Ω—Ç–µ –∫ –∑–∞–¥–∞–Ω–∏—é 2 —Ñ—É–Ω–∫—Ü–∏—è —Å–æ—Å—Ç–æ—è–ª–∞ –±–æ–ª–µ–µ —á–µ–º
–∏–∑ –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö - –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ç–æ–ª—å–∫–æ –ø–µ—Ä–≤—ã–µ –¥–≤–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ.

(ùë•1 - 3)^2 + (ùë•2 ‚àí 1)^2 ‚Üí ùëöùëñùëõ

```python
x = np.arange(1,10,0.05)
y = np.arange(1,10,0.05)

f = np.vectorize(lambda x,y: (x - 3)**2 + (y - 1)**2)
z = f(x,y)

ax = plt.figure().add_subplot(111, projection='3d').plot(x, y, z, color='red')
plt.title('–§—É–Ω–∫—Ü–∏—è (x-3)^2 + (y-1)^2')
plt.show()
```

## –ó–∞–¥–∞–Ω–∏–µ 7

- –ò—Å–ø–æ–ª—å–∑—É—è –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∏ (subplots) –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —Å–µ—Ç–∫—É 2x2 –∏–∑ 4 –≥—Ä–∞—Ñ–∏–∫–æ–≤.
- –£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ —Ä–∞–∑–º–µ—Ä —Å–µ—Ç–∫–∏. 
- –î–æ–±–∞–≤—å—Ç–µ –∑–∞–≥–æ–ª–æ–≤–æ–∫ —Å–µ—Ç–∫–∏, —É–∫–∞–∂–∏—Ç–µ —Ä–∞–∑–º–µ—Ä —à—Ä–∏—Ñ—Ç–∞. 
    - –í –ª–µ–≤–æ–º –≤–µ—Ä—Ö–Ω–µ–º —É–≥–ª—É —Å–µ—Ç–∫–∏ —Ä–∞–∑–º–µ—Å—Ç–∏—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –∏–∑ –ø—É–Ω–∫—Ç–∞ 2 —Ç–µ–∫—É—â–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è, 
    - –≤ –ø—Ä–∞–≤–æ–º –≤–µ—Ä—Ö–Ω–µ–º - –∏–∑ –ø—É–Ω–∫—Ç–∞ 3. 
    - –í –ª–µ–≤–æ–º –Ω–∏–∂–Ω–µ–º - –∫—Ä—É–≥–æ–≤—É—é –¥–∏–∞–≥—Ä–∞–º–º—É –∏–∑ –ø—É–Ω–∫—Ç–∞ 5. 
    - –í –ø—Ä–∞–≤–æ–º –Ω–∏–∂–Ω–µ–º - —Ç—Ä–µ—Ö–º–µ—Ä–Ω—ã–π–≥—Ä–∞—Ñ–∏–∫ –∏–∑ –ø—É–Ω–∫—Ç–∞ 6. 
- –î–ª—è –ª—é–±–æ–≥–æ –∏–∑ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —É—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ —Ü–≤–µ—Ç —Ñ–æ–Ω–∞.

```python
def show_grafics(fig):
    plt.subplots_adjust(left=0.13, right=0.93, top=1.0, bottom= 0.27, wspace= 0.3, hspace=0.3)

    ################ 1 —Ä–∏—Å—É–Ω–æ–∫ #######################
    x = np.arange(1,6,0.05)
    y = np.vectorize(lambda x: math.log(math.sqrt(x)))
    y_x = y(x)

    fig.add_subplot(221).set_facecolor('#AFAFAF')
    plt.plot(x,y_x)
    plt.title('–§—É–Ω–∫—Ü–∏—è ln(sqrt(x))')

    ################ 2 —Ä–∏—Å—É–Ω–æ–∫ #######################
    fig.add_subplot(222)
    plt.plot(x,y_x, marker='o', color=(0.1, 0.2, 0.5))
    plt.grid(visible=True,alpha=0.25,color=(0.3,0.2,0.4))
    plt.title('–§—É–Ω–∫—Ü–∏—è ln(sqrt(x))')

    ################ 3 —Ä–∏—Å—É–Ω–æ–∫ #######################
    fig.add_subplot(223)
    plt.pie(
        x=list(info.values()), 
        labels=list(info.keys()),
        autopct='%1.1f%%',
        colors=['red','orange','yellow','green']
    )
    plt.title('–ö—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')

    ################ 4 —Ä–∏—Å—É–Ω–æ–∫ #######################
    fig.add_subplot(224,projection='3d')
    x = np.arange(1,10,0.05)
    y = np.arange(1,10,0.05)
    f = np.vectorize(lambda x,y: (x - 3)**2 + (y - 1)**2)
    z = f(x,y)
    plt.plot(x, y, z, 'green')
    plt.title('–§—É–Ω–∫—Ü–∏—è (x-3)^2 + (y-1)^2')

    plt.show()
```

```python
show_grafics(plt.figure(figsize=[20,15]))
```
